```python
import pandas as pd
import requests
import os
from datetime import datetime

# -------------------------------
# 1. CONFIG
# -------------------------------
API_KEY = "2c756e7f9e35107c881ab434f7cf06d7" 
BASE_URL = "http://api.openweathermap.org/data/2.5/weather"
CITIES_FILE = "../data/raw/cities_master.csv"  # 100 cities
RAW_FILE = "../data/raw/raw_weather_data.csv"
CLEANED_FILE = "../data/processed/clean_weather_data.csv"
FINAL_FILE = "../data/processed/climate_dashboard_dataset.csv"

# Ensure folders exist
os.makedirs("../data/raw", exist_ok=True)
os.makedirs("../data/processed", exist_ok=True)
os.makedirs("../data/processed", exist_ok=True)


# -------------------------------
# 2. Fetch Data
# -------------------------------
def fetch_data():
    print("Fetching data from OpenWeather API...")

    cities_df = pd.read_csv(CITIES_FILE)
    data_list = []

    for city in cities_df["City"]:
        params = {
            "q": city,
            "appid": API_KEY,
            "units": "metric"
        }
        response = requests.get(BASE_URL, params=params).json()

        # Check for valid response
        if response.get("main"):
            data_list.append({
                "City": city,
                "Temperature": response["main"]["temp"],
                "Humidity": response["main"]["humidity"],
                "Pressure": response["main"]["pressure"],
                "Datetime": datetime.now()
            })

    global weather_df
    weather_df = pd.DataFrame(data_list)

    # Save raw data
    weather_df.to_csv(RAW_FILE, index=False)
    print(f"Raw data saved: {RAW_FILE}")


# -------------------------------
# 3. Clean Data
# -------------------------------
def clean_data():
    print("Cleaning data...")

    df = pd.read_csv(RAW_FILE)
    df["Datetime"] = pd.to_datetime(df["Datetime"], errors="coerce")

    # Missing values
    df["Temperature"].fillna(df["Temperature"].mean(), inplace=True)
    df["Humidity"].fillna(df["Humidity"].median(), inplace=True)
    df["Pressure"].fillna(method="ffill", inplace=True)

    # Outliers (IQR capping)
    for col in ["Temperature", "Humidity", "Pressure"]:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        df[col] = df[col].clip(lower, upper)

    # Load old cleaned data if exists
    if os.path.exists(CLEANED_FILE):
        old_df = pd.read_csv(CLEANED_FILE)
        old_df["Datetime"] = pd.to_datetime(old_df["Datetime"], errors="coerce")
    else:
        old_df = pd.DataFrame()

    # Merge old + new
    combined_df = pd.concat([old_df, df], ignore_index=True)

    # Remove duplicates (City + Datetime)
    combined_df.drop_duplicates(subset=["City", "Datetime"], keep="last", inplace=True)

    # Save cleaned data
    combined_df.to_csv(CLEANED_FILE, index=False)
    print(f"Cleaned data saved: {CLEANED_FILE}")


# -------------------------------
# 4. Aggregate Data for Dashboard
# -------------------------------
def aggregate_data():
    print("Aggregating data for dashboard...")

    df = pd.read_csv(CLEANED_FILE)
    df["Date"] = pd.to_datetime(df["Datetime"]).dt.date

    agg_df = df.groupby(["City", "Date"]).agg({
        "Temperature": "max",
        "Humidity": "mean",
        "Pressure": "mean"
    }).reset_index()

    agg_df.to_csv(FINAL_FILE, index=False)
    print(f"Aggregated data saved: {FINAL_FILE}")


# -------------------------------
# 5. Main Pipeline
# -------------------------------
def run_pipeline():
    fetch_data()
    clean_data()
    aggregate_data()
    print("✅ Full pipeline executed successfully!")


# -------------------------------
# 6. Execute
# -------------------------------
if __name__ == "__main__":
    run_pipeline()
```
